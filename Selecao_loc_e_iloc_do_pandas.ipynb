```
import pandas as pd

#chamando o dataset
df = pd.read_csv("https://raw.githubusercontent.com/matheusbattisti/ds3_loc_iloc/master/nba_draft_2015.csv")

df.head()
```

```
df.shape
```

# **LOC**

```
# loc com índice
df.loc[[5]]
```

```
# loc com array
df.loc[[0,1,2]]
```

```
# loc com um "pedaço" do dataframe
# Nesse caso, não é da quarta posição (índice 3) até a oitava posição (indice 7), e sim do índice de número quatro até o índice de número oito
df.loc[4:8]
```

```
# Setando um index para podermos trabalhar com o nome das linhas
df2 = df.set_index('Player')

df2.head()
```

```
# Agora podemos chamar pelo índice com nome, por exemplo:
df2.loc['Justise Winslow']
```

```
# Selecionar apenas as colunas que queremos também é possivel, neste exemplo um array com as linhas e outro de colunas
df2.loc[['Karl-Anthony Towns', 'Stanley Johnson', 'D`Angelo Russell'], ['Position', 'ID']]
```

```
# Localizar os atletas com probabilidade maior ou igual a 10% de serem superstars
# Multiplicampos por 100, para ficar mais facil o entendimento

df.loc[(df['Superstar']*100) >= 10]
```

```
# Mudando os 5 primeiros registros para o ano de 2018
# Nesse caso, a seleção das linhas não pega o limite inferior, no caso, o 5, pega apenas os 5 primeiros e não da 0 a 5 posição
df2.loc[0:5, 'Draft Year'] = 2018

df2.head(5)
```

# **ILOC**

```
# Selecionar a primeira linha do dataset
df.iloc[0]
```

```
# Selecionar a última linha
df.iloc[-1]
```

```
# Todos os dados da primeira coluna do dataset
df.iloc[:,0]
```

```
# Slice: Do primeiro ao quinto dado da última coluna
# Nesse caso, a seleção das linhas não pega o limite inferior, no caso, o 5, pega apenas os 5 primeiros e não da 0 a 5 posição
df.iloc[0:5,-1]
```

```
# Array: Do primeiro ao quinto dado da última coluna
# Com o array a seleção das linhas não pega os 6 valores da 0 a 5 posição
df.iloc[[0,1,2,3,4,5],-1]
```

```
# resgatando as primeiras três linhas do dataset
# Nesse caso, a seleção das linhas não pega o limite inferior, no caso, o 3, pega apenas os 3 primeiros e não da 0 a 3 posição
df.iloc[0:3]
```

```
# Todos os dados da segunda e terceira coluna
# Para coluna também a seleção das linhas não pega o limite inferior, no caso, o 3. Pega apenas as duas colunas especificadas, a segunda (coluna na posição 1) e 
# a terceira (coluna na posição 2), e não das colunas de 1 a 3 posições
df.iloc[:, 1:3]
```

```
# 1º,3º e 5º elementos e 6ª a 8ª colunas
# Para coluna também a seleção das linhas não pega o limite inferior, no caso, o 8. Pega apenas as três colunas especificadas, a quinta (coluna na posição 4) e 
# a sétima (coluna na posição 6), e não das colunas de 5 a 8 posições

df.iloc[[0,2,4], 5:8]
```

```
# quando solicitamos a o iloc uma linha ele retorna um Pandas Series, já múltiplas retorna um Pandas DataFrame.
# O pandas fornece duas maneiras de estruturar os nossos dados, o DataFrame que possui uma estrutura tabular de N-dimensões, onde cada coluna é um campo da 
# tabela e cada linha um registro, e a Series, que é uma matriz unidimensional que contém uma sequencia de valores acompanhado de seus respectivos índices.

print(type(df.iloc[0]))
print(type(df.iloc[0:3]))
```
